```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train_stats.csv', index_col=0)

data.head()
```

```python
loss_values = data['loss_values']
test_loss_values = data['test_loss_values']
epoch_count = data['epoch_count']
```

```python
plt.semilogy()
plt.plot(epoch_count, loss_values, label='Training Loss')
plt.plot(epoch_count, test_loss_values, label='Test Loss')
plt.legend()

plt.show()
```
